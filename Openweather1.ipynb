```python
import datetime
import json
import urllib.request

def time_converter(time):
    converted_time = datetime.datetime.fromtimestamp(
        int(time)
    ).strftime('%I:%M %p')
    return converted_time

def url_builder1(city):
    user_api = '469d1d7509c8b5a4b70cf03a920b87c7'  # metemos el API key de nuestro usuario de la web
    unit = 'metric'  #Se elige la unidad métrica con la que se va a trabajar (esto depende de las unidades en las que se obtengan los datos de temperatura)-->For Fahrenheit use imperial, for Celsius use metric, and the default is Kelvin.
    api = 'http://api.openweathermap.org/data/2.5/weather?q='     # Search for your city ID here: http://bulk.openweathermap.org/sample/city.list.json.gz
    full_api_url = api + str(city) + '&mode=json&units=' + unit + '&APPID=' + user_api
    print("\n URL de la API:", full_api_url)
    return full_api_url

def data_fetch(full_api_url): #funcion que se encarga de extraer datos de nuestra web, a la cual se le pasa el url de la api que se quiere acceder
    url = urllib.request.urlopen(full_api_url)
    output = url.read().decode('utf-8')
    raw_api_dict = json.loads(output)
    url.close()
    #print(raw_api_dict)
    return raw_api_dict

if __name__ == '__main__':
    try:
        now=datetime.datetime.now() #nos devuelve el current local datetime, osea la fecha y hora correspondiente a la ciudad especificada
        datos=data_fetch(url_builder1('Alicante,ES')) #se llama a data_fetch que nos obteienlos datos de full_api_url correspondiente a la ciudad especificada, en este caso ALicante
        
        print("\n Datos obtenidos de la API:", datos) #se imprimen los datos obtenidos por data_fetch
        print(now)  #se imprime el current local datetime
       
        print("\n Datos obtenidos del main:")
        for key in datos['main']: 
          print(key, ":", datos['main'][key]) #se imprimen los datos que nos ofrece main (temp, feels_like, temp_min,temp_max, pressure, sea_level,)    

        print("\n Datos obtenidos de wind:")
        for key in datos['wind']:
          print(key, ":", datos['wind'][key]) #se imprimen los datos que ofrece la etiqueta wind (speed, deg, gust)
        
        print("\n Datos obtenidos de clouds (porcentaje de nubosidad):")#se imprimen los datos que ofrece la etiqueta clouds (all)
        for key in datos['clouds']:
          print(key, ":", datos['clouds'][key])

        print("\n Datos obtenidos de weather:") #se imprimen los datos que ofrece la etiqueta weather (ID-->weather condition id,Main--> group of weather parameters (Rain, Snow, Extreme etc.),Description-->weather condition within the group, Icon-->weather icon id)
        print(datos['weather'][0])

        
        print ("\n Hora a la que se ha obtenido la predicción de los datos:", datos['dt'])


        for key in datos['sys']:
          print(key, ":", datos['sys'][key])
        
        print("\n Hora convertida:",time_converter(1612681284))
        print("\n Hora a la que amanece:", time_converter(datos['sys']['sunrise']))
        print("\n Hora a la que anochece:", time_converter(datos['sys']['sunset']))
    except IOError:
        print('error')

```
